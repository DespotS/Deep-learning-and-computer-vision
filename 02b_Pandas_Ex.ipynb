We now want to look at the overall distribution of **class levels** of these high school students. Unfortunately, the column is of type `object`, so we can not calcuate any basic statistics based on that. To do that, we would need to add a new column with the numeric representation of the class level.
* <div style="color:blue">Find out, which values are contained in the column <code>Class Level</code>. You want to make sure, that every value is properly is considered for transformation. You could use the following pandas function to see all unique values: <a href="https://pandas.pydata.org/docs/reference/api/pandas.unique.html">pd.unique</a>.</div>
* <div style="color:blue">Add a new column <code>class_level_numeric</code> to the dataframe, containing only the numeric value of the high school year of each student. This can be done in numerous ways. The most robust is probably using the <a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.map.html">pandas map function</a>. You could also use the <a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html">replace function</a> or find another clever way based on the values contained in this column.</div>
* <div style="color:blue">Make sure that the <code>Dtype</code> of your new column is an <code>integer</code>.</div>
* <div style="color:blue">Finally, use the <code>describe</code> function to check the statistical information of the new column.</div>

```python
# Your solution goes here:
import pandas as pd

clipdf = pd.read_clipboard()
unique_values = pd.unique(clipdf['Class Level'])
print("Unique values in 'Class Level':", unique_values)

class_level_map = {
    '1. Freshman': 1,
    '2. Sophomore': 2,
    '3. Junior': 3,
    '4. Senior': 4
}
clipdf['class_level_numeric'] = clipdf['Class Level'].map(class_level_map)

clipdf['class_level_numeric'] = clipdf['class_level_numeric'].astype(int)

print(clipdf['class_level_numeric'].describe())
```

Now that we have a numeric representation of the class level, we can use that to take a more detailed look into its distribution for specific subgroups.
* <div style="color:blue">Use a <b>pivot table</b> to answer the following question: What is the median and standard deviation of <code>class_level_numeric</code> for each <code>Major</code>, separated by <code>Gender</code>? Use <code>Gender</code> as the first index of you multi-index pivot table!</div>
* <div style="color:blue">Now use a filter on your pivot table to only show median values of less than 2. <b>Beware of the multi-index!</b></div>

```python
# Your solution goes here:
import pandas as pd

clipdf = pd.read_clipboard()
unique_values = pd.unique(clipdf['Class Level'])
print("Unique values in 'Class Level':", unique_values)

class_level_map = {
    '1. Freshman': 1,
    '2. Sophomore': 2,
    '3. Junior': 3,
    '4. Senior': 4
}
clipdf['class_level_numeric'] = clipdf['Class Level'].map(class_level_map)

clipdf['class_level_numeric'] = clipdf['class_level_numeric'].astype(int)

print(clipdf['class_level_numeric'].describe())

pivot_table = clipdf.pivot_table(
    index = ['Gender','Major'],
    values= 'class_level_numeric',
    aggfunc = ['median', 'std'])

f_pivot = pivot_table[pivot_table[('median', 'class_level_numeric')]<2]
print(f_pivot)
```
